```python
import pandas as pd

# The URL to the raw Titanic CSV file hosted on GitHub
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Tell Pandas to read that URL and turn it into a DataFrame
titanic_df = pd.read_csv(url)

# Display the first 5 rows of the dataset
titanic_df.head()
```

```python
# Isolate the 'Age' column and calculate the mean (average)
average_age = titanic_df['Age'].mean()

# Print the result nicely rounded to 2 decimal places
print(f"The average age on the Titanic was: {average_age:.2f} years")
```

```python
# 1. Group the passengers by 'Sex'
# 2. Look at the 'Survived' column (which is just 1s and 0s)
# 3. Calculate the mean (average) for each group
# 4. Multiply by 100 to make it a clean percentage

survival_rates = titanic_df.groupby('Sex')['Survived'].mean() * 100

print("Survival Rates by Gender:")
print(survival_rates.round(2).astype(str) + '%')
```

```python
# Import matplotlib, the standard Python graphing tool
import matplotlib.pyplot as plt

# 1. Take our calculated rates and tell Pandas to make a 'bar' chart
# 2. We'll add some custom colors: soft red for female, soft blue for male
survival_rates.plot(kind='bar', color=['#ff9999', '#66b3ff'], edgecolor='black')

# 3. Add professional labels and a title
plt.title('Titanic Survival Rates by Gender', fontsize=14, fontweight='bold')
plt.ylabel('Survival Rate (%)', fontsize=12)
plt.xlabel('Gender', fontsize=12)

# 4. Keep the 'female' and 'male' text flat (rotation=0) instead of sideways
plt.xticks(rotation=0)

# 5. Display the final chart
plt.show()
```
